```python
import pandas as pd
```

```python
salary = pd.read_csv('decision_tree\salaries.csv')
salary
```

```python
x = salary[['company','job','degree']]
```

```python
y = salary.salary_more_then_100k
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_company = LabelEncoder()
encoded_job = LabelEncoder()
encoded_degree = LabelEncoder()
```

```python
x['enc_company'] = encoded_company.fit_transform(x['company'])
x['enc_job'] = encoded_job.fit_transform(x['job'])
x['enc_degree'] = encoded_degree.fit_transform(x['degree'])
x
```

```python
x.drop(['job','company','degree'],axis=1,inplace=True)
x
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```
